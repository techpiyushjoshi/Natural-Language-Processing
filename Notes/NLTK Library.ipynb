## Tokenization

```python
from nltk.tokenize import sent_tokenize
```

```python
text = "A sentence is a set of words that is complete in itself, typically containing a subject and predicate, conveying a statement, question, exclamation, or command, and consisting of a main clause and sometimes one or more subordinate clauses."
```

## Normalization

```python
text = text.lower()
```

```python
text
```

```python
sent_tokenize(text)
```

```python
from nltk.tokenize import word_tokenize 
```

```python
tokens = word_tokenize(text)
```

```python
print(tokens)
```

## Frequency Distribution

```python
from nltk.probability import FreqDist
```

```python
fdist = FreqDist(tokens)
```

```python
fdist
```

```python
print(dict(fdist))
```

```python
import matplotlib.pyplot as plt
```

```python
fdist.plot(20,cumulative=False)
plt.show()
```
