```python
text="""the nervous system and brain function do not often associate with infectious diseases, but numerous infectious agents affect the brain and peripheral neurons in deleterious ways. infectious agents often behave differently in the brain than in the circulation or other organs, complicating matters for scientists working to stop neurotropic infectious diseases"""
```

## Normalization

```python
text=text.lower()
```

## Tokenization

<p>Tokenization is the first step in text analytics. The process of breaking down text paragraphs into smaller chunks such as words or sentences is called Tokenization. </p>

```python
from nltk.tokenize import sent_tokenize
```

```python
sents=sent_tokenize(text)

print(sents)
```

```python
from nltk.tokenize import word_tokenize

tokens=word_tokenize(text)

print(tokens)
```

## Frequency Distribution

```python
from nltk.probability import FreqDist

fdist = FreqDist(tokens)

fdist
```

```python
print(dict(fdist))
```

```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt

fdist.plot(20,cumulative=False)
plt.show()
```

## Removing Punctuations and Stopwords

```python
print(tokens)
```

```python
import string

# punctuations
punctuations=list(string.punctuation)

# stopwords
stop_words=list(stopwords.words("english"))

# Custom keywords
custom_list=[]

tokens_p=[]

for i in tokens:
    if i not in punctuations+stop_words+custom_list:
        tokens_p.append(i)
```

```python
tokens_p
```

## Lexicon Normalization

### Stemming

```python
# Stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()

stemmed_words=[]

for w in tokens_p:     
     stemmed_words.append(ps.stem(w))

print("Filtered Sentence:",tokens_p)
print("Stemmed Sentence:",stemmed_words)
```

### Lemmatization

```python
from nltk.stem.wordnet import WordNetLemmatizer

lem = WordNetLemmatizer()
            
lemmatized_words=[]

for w in tokens_p:     
     lemmatized_words.append(lem.lemmatize(w,"n"))

print("Filtered Words:",tokens_p)
print("lemmatized Words:",lemmatized_words)
```

### Difference between stemming and lemmatization.

```python
#Lexicon Normalization#performing stemming and Lemmatization
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer

lem = WordNetLemmatizer()
stem = PorterStemmer()
word = "flying"

print("Lemmatized Word:",lem.lemmatize(word,"v"))

print("Stemmed Word:",stem.stem(word))
```

## PoS Tags

```python
sent = "Albert Einstein was born in Ulm, Germany in 1879."

tokens_=nltk.word_tokenize(sent)

pos_=nltk.pos_tag(tokens_)

print(pos_)
```

## NER: Named Entity Recognition

```python
from nltk import ne_chunk

sent="New York City on Tuesday declared a public health emergency and ordered mandatory measles vaccinations amid an outbreak, becoming the latest national flash point over refusals to inoculate against dangerous diseases."
```

```python
for chunk in ne_chunk(nltk.pos_tag(word_tokenize(sent))):
        if hasattr(chunk, 'label'):
            print(chunk.label(), ' '.join(c[0] for c in chunk))
```
